## Lexique

* R0 
  * abattement forfaitaire du bar√®me locatif ordinaire 
  * param√®tre de ressources > revaloris√© au 1er janvier
  * depuis le 1er janvier 2021, le R0 est revaloris√© selon l‚Äôindice des prix √† la consommation
hors tabac (IPC) de l‚Äôann√©e N-1, son √©volution √©tant calcul√©e sur le mois d‚Äôoctobre

* APL 
  * montant mensuel (dans les formules, c'est le montant avant √©tapes de finalisation)

* L 
  * loyer mensuel r√©el pris en compte dans la limite d‚Äôun plafond variable en fonction de trois zones g√©ographiques et du nombre de personnes √† charge.

* L0 
  * le loyer minimal (ou la mensualit√© de remboursement minimale) que le m√©nage doit consacrer √† son logement. 
  * Il est fonction du revenu et du nombre de personnes √† charge.

* E 
  * l‚Äô√©quivalence de loyer et de charges locatives prise en compte dans la limite d‚Äôun plafond variable dans les m√™mes conditions que L.

* E0 
  * l‚Äô√©quivalence de loyer et de charges locatives **minimale** que le m√©nage doit consacrer √† son logement 
  * elle est fonction du revenu et du nombre de personnes √† charge

* C 
  * le forfait de charges
  * Il est fonction de la composition familiale.

* Pp
  * la participation personnelle du m√©nage √† la d√©pense de logement.
  * elle est fonction de la composition familiale, du loyer et des ressources.

* K 
  * le coefficient de prise en charge de la diff√©rence entre la mensualit√© de remboursement r√©elle plafonn√©e, major√©e du forfait de charges (...) ou l‚Äô√©quivalence de loyer et de charges locatives en APL foyer, et le loyer minimal (ou l‚Äô√©quivalence de loyer et de charges locatives minimale). 
  * Il est fonction du revenu et du nombre de personnes √† charge.

> Source : [Brochure bar√®me APL 2024 (PDF)](https://www.ecologie.gouv.fr/sites/default/files/documents/Brochure-bareme-2024-APL.pdf)

## Secteur locatif ordinaire

En secteur locatif ordinaire (c‚Äôest-√†-dire hors foyers), les bar√®mes APL et AL sont identiques depuis le 1er janvier 2001.

Depuis la r√©forme intervenue le 1er janvier 2001, le montant de l‚Äôaide est obtenu par
application de la m√™me formule en AL et en APL (cf. article D. 823-16 du CCH) :
```
APL ou AL = L + C - Pp
```

> Source : [Brochure bar√®me APL 2024 (PDF)](https://www.ecologie.gouv.fr/sites/default/files/documents/Brochure-bareme-2024-APL.pdf)

````python
# APL ou AL = L + C ‚Äì Pp
# APL ou AL = L + C ‚Äì [ P0 + Tp * ( R ‚Äì R0 ) ]
# APL ou AL = L + C ‚Äì [ P0 + ( TF + TL ) * ( R ‚Äì R0 ) ]
L_value + C_value - Pp_value
````

## Secteur foyer

En secteur foyer, il existe deux bar√®mes en APL :
* l‚ÄôAPL 1 foyer qui concerne la plupart des logements-foyers ;
* l‚ÄôAPL 2 foyer qui concerne les foyers de jeunes travailleurs (FJT) et les r√©sidences sociales, existants conventionn√©s sans travaux, √† compter du 1er octobre 1990 pour les FJT et du 1er janvier 1995 pour les r√©sidences sociales.

(...)

Le montant de l‚ÄôAPL foyer est obtenu par application de la formule (cf. article D. 832-24 du CCH) :
```
APL = K * (E ‚Äì E0)
```

Avec application de deux bar√®mes, APL 1 foyer et APL 2 foyer (cf. articles D. 832-25 et D. 832-26 du CCH).
